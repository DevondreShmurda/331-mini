```python
# prompt: can we make the column names, the line of row 2

from google.colab import auth
import gspread
from google.auth import default
import pandas as pd

auth.authenticate_user()

creds, _ = default()
gc = gspread.authorize(creds)

spreadsheet_url = 'https://docs.google.com/spreadsheets/d/1mc72_VV6qVqg6RVoJQVGu5XY81QncEXIn0vGUo2zVs0/edit#gid=2060843025'
workbook = gc.open_by_url(spreadsheet_url)

worksheet = workbook.worksheet('PR-OR') # Access the 'PR-OR' sheet

# Get all values from the worksheet
rows = worksheet.get_all_values()

# Convert to a pandas DataFrame
df = pd.DataFrame.from_records(rows)

# Set the second row as the header
df.columns = df.iloc[1]

# Eliminate rows 0 and 1
df = df.drop([0, 1])

# Print the DataFrame
df
```
